```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
data = pd.read_csv("C:\\Users\\immoh\\Downloads\\Heart_Disease_Prediction.csv")
data.head()
```

```python
data.describe().T
```

```python
sns.pairplot(data)
plt.suptitle("Pairplot of Numerical Features", y = 1.02)
plt.show()
```

```python
plt.figure(figsize = (15,10))
numerical_cols = data.select_dtypes(include = 'number').columns
for i, col in enumerate(numerical_cols):
    plt.subplot(3,5, i+1)
    sns.histplot(data[col], kde = True, color = 'teal')
    plt.title(f'Distribution of {col}')
    
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (15,8))
categorical_cols = data.select_dtypes(include = 'object').columns
for i, col in enumerate(categorical_cols):
    plt.subplot(2,3, i+1)
    sns.histplot(data[col], kde = True, color = 'teal')
    plt.title(f'Distribution of {col}')
    
plt.tight_layout()
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.histplot(data["Age"], bins = 20, kde = True, palette = 'teal')
plt.title("Age Distribution", fontweight = 'bold')
plt.show()
```

```python
sns.boxplot(x = "Heart Disease", y = "Age", data = data)
plt.title("Age vs Heart Disease")
plt.show()
```

```python
sns.countplot(x = "Sex", hue = "Heart Disease", data = data)
plt.title("Sex vs Heart Disease")
plt.show()
```

```python
sns.countplot(x = "Chest pain type", hue = "Heart Disease", data = data)
plt.title("Chest Pain Type vs Heart Disease")
plt.show()
```

```python
sns.boxplot(x = "Heart Disease", y = "BP", data = data)
plt.title("Blood Pressure vs Heart Disease")
plt.show()
```

```python
sns.boxplot(x = "Heart Disease", y = "Cholesterol", data = data)
plt.title("Cholesterol vs Heart Disease")
plt.show()
```

```python
sns.countplot(x = "Thallium", hue = "Heart Disease", data = data)
plt.title("Thallium Test vs Heart Disease")
plt.show()
```

```python
plt.figure(figsize = (14,10))
corr = data.drop("Heart Disease", axis = 1).corr()
sns.heatmap(corr, annot = True, cmap = "coolwarm", fmt = ".2f")
plt.title("Feature Correlation Heatmap", fontweight = 'bold')
plt.show()
```
